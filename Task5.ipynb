```python
#b. Visualization using Seaborn/Matplotlib
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(df[['Rank', 'Score', 'SubmissionCount']])
plt.show()
```

```python
sns.heatmap(df[['Rank', 'Score', 'SubmissionCount']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# c. Identify Relationships & Trends
```

```python
sns.scatterplot(x='Rank', y='Score', data=df)
plt.title('Rank vs Score')
plt.show()
```

```python
sns.boxplot(x='SubmissionCount', y='Score', data=df)
plt.title('Score Distribution by Submission Count')
plt.show()
```

```python
# d. Histograms, Boxplots, Scatterplots
```

```python
df['Score'].hist(bins=15, edgecolor='black')
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(y='Score', data=df)
plt.title('Score Boxplot')
plt.show()
```
